```python
# 데이터 로드
df = pd.read_csv("../data/heart.csv")

# 기본 정보 확인
print("=== 데이터 shape ===")
print(df.shape)

print("\n=== 처음 5행 ===")
display(df.head())

print("\n=== 데이터 타입 및 결측값 ===")
display(df.info())

print("\n=== 기술통계 ===")
display(df.describe())
```

```python
# 연속형 특성만 선택
continuous_cols = ["age", "trestbps", "chol", "thalach", "oldpeak"]

plt.figure(figsize=(15, 5))
for i, col in enumerate(continuous_cols):
    plt.subplot(1, 5, i+1)
    sns.boxplot(y=df[col], palette="Set2")
    plt.title(col)
plt.suptitle("연속형 특성 이상치 확인 (Boxplot)")
plt.tight_layout()
plt.savefig("../data/boxplot_outliers.png")
plt.show()
print("그래프 저장 완료")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# 연속형 / 범주형 컬럼 정의
continuous_cols = ["age", "trestbps", "chol", "thalach", "oldpeak"]
categorical_cols = ["sex", "cp", "fbs", "restecg", "exang", "slope", "ca", "thal"]

# 연속형 파이프라인: 결측값 → 중앙값 대치 → 스케일링
continuous_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

# 범주형 파이프라인: 결측값 → 최빈값 대치
categorical_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent"))
])

print("파이프라인 정의 완료")
print(f"연속형 특성: {continuous_cols}")
print(f"범주형 특성: {categorical_cols}")
```

```python
from sklearn.compose import ColumnTransformer

# 전체 전처리 파이프라인
preprocessor = ColumnTransformer([
    ("continuous", continuous_pipeline, continuous_cols),
    ("categorical", categorical_pipeline, categorical_cols)
])

# 타깃 분리
X = df.drop("target", axis=1)
y = df["target"]

# 파이프라인 적용 확인 (fit_transform은 나중에 train에서만!)
print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
print(f"타깃 분포:\n{y.value_counts()}")
```
